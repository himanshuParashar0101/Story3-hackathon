```python
import requests

# Base URL for the API
base_url = 'https://story3.com/api/v2'

# Headers for authentication
headers = {
    'accept': 'application/json',
    'x-auth-token': 'Your_API_Key_Here',  # Replace with your actual API key
}

# URL for creating a new story
stories_url = f'{base_url}/stories'

# New story data for Story 12
new_story_data = {
    'title': 'The Rajput Legacy',
    'body': (
        "In the majestic landscapes of Rajasthan, 'The Rajput Legacy' unfolds the saga of a young warrior determined to reclaim "
        "the lost glory of their Rajput lineage. Ancient forts echo with the tales of valor, and the Thar Desert becomes a vast canvas "
        "where the warrior's destiny is written in the sands of time.\n\n"
        "As the warrior embarks on a quest, they encounter challenges that test not only their combat skills but also their principles. "
        "'The Rajput Legacy' becomes a journey of self-discovery, honor, and the enduring spirit of Rajputana."
    ),
    'tags': ['Rajasthan', 'Rajput Heritage', 'Warrior', 'Self-Discovery'],
}

# Make a POST request to create the new story
response = requests.post(stories_url, headers=headers, json=new_story_data)

if response.status_code == 201:
    new_story_info = response.json()
    selected_story_hash = new_story_info['hashId']
    print(f'New Story created successfully: {new_story_info}')
else:
    print(f'Error creating a new story. Status code: {response.status_code}')
    print(f'Error response: {response.json()}')

# URL for creating a new twist
twists_url = f'{base_url}/twists'

# First twist data for Story 12
first_twist_data = {
    'hashId': selected_story_hash,
    'hashParentId': selected_story_hash,
    'body': (
        "In the heart of Rajasthan, the warrior discovers an ancient relic said to hold the key to unlocking the dormant powers of "
        "Rajput warriors. 'The Rajput Legacy' transforms into a mystical journey, blending martial prowess with ancient mysticism as "
        "the warrior delves into the secrets of the relic."
    ),
    'title': 'Relic of Rajput Valor',
    'tags': ['Ancient Relic', 'Mystical Powers', 'Rajput Warriors', 'Journey of Discovery'],
}

# Make a POST request to create the first twist
response = requests.post(twists_url, headers=headers, json=first_twist_data)

if response.status_code == 201:
    first_twist_info = response.json()
    print(f'First Twist created successfully: {first_twist_info}')
else:
    print(f'Error creating the first twist. Status code: {response.status_code}')
    print(f'Error response: {response.json()}')

# Second twist data for Story 12
second_twist_data = {
    'hashId': selected_story_hash,
    'hashParentId': first_twist_info['hashId'],
    'body': (
        "As the warrior unveils the relic's powers, they are faced with a choice â€” to use the newfound abilities for personal gain or "
        "to safeguard the honor and traditions of the Rajput lineage. 'The Rajput Legacy' becomes a tale of moral dilemmas, forging a "
        "path that preserves the legacy of Rajasthan's warriors."
    ),
    'title': 'Dilemmas of Honor',
    'tags': ['Moral Dilemmas', 'Legacy Preservation', 'Rajput Honor', 'Choices and Consequences'],
}

# Make a POST request to create the second twist
response = requests.post(twists_url, headers=headers, json=second_twist_data)

if response.status_code == 201:
    second_twist_info = response.json()
    print(f'Second Twist created successfully: {second_twist_info}')
else:
    print(f'Error creating the second twist. Status code: {response.status_code}')
    print(f'Error response: {response.json()}')
```
